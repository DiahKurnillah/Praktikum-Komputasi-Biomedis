Dengan demikian, adanya komputasiyangmemanfaatkan phyton dapat mempermudah pekerjaan manusia dalam menentukan interpolasi berdasarkan data yang diberikan.
Interpolasi Lagrange digunakan untuk mencari beberapa titik penghubung dari n titik $P1(x1, y1), P2(x2, y2), P3(x3, y3),..., PN(xN, yN)$ dengan menggunakan aproksimasi fungsi polinomial yang tersusun dalam kombinasi deret dan didefinisikan sebagai berikut.
$$y = \sum_{i=1}^{N} y_i \prod_{j \neq i} \frac{(x - x_j)}{(x_i - x_j)}
$$


# Preliminary Task
Buatlah interpolasi Lagrange dari data berikut ini: (0,1), (1,2), (3,4), (6,-1)

```python
import numpy as np
import pylab

x = [0.0, 1.0, 3.0, 6.0]
y = [1.0, 2.0, 4.0, -1.0]
hasil = np.poly1d([0.0])
n = len(x)

for i in range (0,n):
  num = np.poly1d([1.0])
  denum = 1.0
  for j in range (0,n):
    if (j!=i):
      num = num*np.poly1d([1.0,-x[j]])
      denum = denum*(x[i]-x[j])
  hasil = hasil + y[i]*(num/denum)
print (hasil)

xcoba = 20
ycoba = hasil(xcoba)
print ('y coba =',ycoba)

x1 = np.arange(min(x)-5,max(x)+5,0.1)
y1 = hasil(x1)

pylab.plot(x1,y1,label='Interpolasi Lagrange')
pylab.plot(x,y,'x', label='Data Asli')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True)

x2 = np.array([-2.0,1.5,4.7,8.2])
y2 = hasil(x2)
pylab.plot(x2,y2,'-o',label='Titik Evaluasi')
pylab.legend()
```

Pertama-tama, kita mendeklarasikan dua array, x dan y, yang berisi data nilai x(n) dan y(n) dari tabel data. Kemudian, kita mendeklarasikan variabel hasil yang akan menampung persamaan interpolasi menggunakan sintaks poly1d dari modul numpy. Nilai awal hasil diatur menjadi 0.
Variabel n diinisialisasi dengan jumlah data pada array x menggunakan fungsi len(x), sehingga n akan menjadi jumlah data pada x (dalam contoh ini, n = 4).
Selanjutnya, kita menggunakan dua loop bersarang. Loop pertama (looping variabel i) digunakan untuk menghasilkan persamaan polinomial akhir. Di dalam loop pertama, ada loop kedua (looping variabel j) yang digunakan untuk menghitung nilai num dan denum.
Variabel num dan denum digunakan untuk mewakili nilai (x - xj) dan (xi - xj) pada persamaan umum interpolasi Lagrange sebagai pembilang dan penyebut. Nilai awal num diisi dengan 1.0, dan denum juga diisi dengan 1.0.
Dalam loop kedua, kita menggunakan blok if untuk memastikan bahwa nilai i tidak sama dengan nilai j agar program bisa melanjutkan perhitungan. Di dalam blok if, kita menghitung nilai num dan denum dengan mengalikan nilai num dan denum sebelumnya dengan faktor (x - xj) dan (xi - xj).
Variabel hasil digunakan untuk memproses hasil dari loop kedua ke dalam loop pertama. Nilai hasil diupdate dengan menjumlahkan nilai hasil sebelumnya dengan nilai y saat i yang dikalikan dengan hasil bagi num oleh denum. Ini sesuai dengan rumus interpolasi Lagrange.
Hasil akhir dari loop pertama (hasil) adalah persamaan polinomial interpolasi Lagrange dengan pangkat 3, karena kita memiliki 4 data pada array x. Kemudian, kita melakukan uji coba persamaan interpolasi tersebut dan menampilkan plotnya menggunakan pylab.

1.	Tentukan urutan interpolasi polinomial Lagrange yang tepat melewati tujuh titik yang ditunjukkan di atas dan Plotkan fungsinya!


```python
import numpy as np
import pylab
from prettytable import PrettyTable

t = PrettyTable()
t.field_names = ['x','y','f(x)']
x = np.array([0.1,0.3,0.5,0.7,0.9,1.1,1.3])
y = np.array([0.030,0.067,0.148,0.248,0.320,0.518,0.697])
hasil=np.poly1d([0.0])
n=len(x)
y2=np.zeros(n,float)

for i in range(0,n):
  num = np.poly1d([1.0])
  denum = 1.0
  for j in range(0,n):
    if i!=j:
      num = num*np.poly1d([1.0,-x[j]])
      denum = denum*(x[i]-x[j])
  hasil = hasil+y[i]*(num/denum)
  y2[i] = hasil(x[i])
  t.add_row([x[i],y[i],(round(y2[i],3))])
  t.align='c'

print('Persamaan Interpolasinya adalah f(x)=\n',hasil)
print("\nTabel Hasil")
print (t)

order_interpolation = n - 1
x1 = np.arange(min(x),max(x)+0.1,0.1)
y1 = hasil(x1)
pylab.plot(x1,y1,'-o',label=f'Interpolasi Lagrange Urutan {order_interpolation}')
pylab.plot(x,y,'*',label='Data Titik')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.grid(True)
pylab.title('Interpolasi Polinomial Lagrange')
pylab.legend()
```

Terlihat bahwa nilai data awal masih masuk dalam rentang interpolasi. Untuk menguji persamaan dapat dilakukan substitusi nilai x dalam persamaan f(x)

# Task 3
Apa yang terjadi jika nilai x tidak berada dalam rentang data interpolasi Lagrange? Jelaskan jawaban Anda dan tunjukkan dalam sebuah plot!

```python
import numpy as np
import matplotlib.pyplot as plt
from prettytable import PrettyTable

t = PrettyTable()
t.field_names = ['x', 'y', 'f(x)']

x = np.array([0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3])
y = np.array([0.030, 0.067, 0.148, 0.248, 0.320, 0.518, 0.697])

n = len(x)
hasil = np.poly1d([0.0])
y2 = np.zeros(n, float)

for i in range(0, n):
    num = np.poly1d([1.0])
    denum = 1.0
    for j in range(0, n):
        if i != j:
            num = num * np.poly1d([1.0, -x[j]])
            denum = denum * (x[i] - x[j])
    hasil = hasil + y[i] * (num / denum)
    y2[i] = hasil(x[i])
    t.add_row([x[i], y[i], round(y2[i], 3)])
    t.align = 'c'

order_interpolation = n - 1

print('Persamaan Interpolasinya adalah f(x) =\n', hasil)
print("\nTabel Hasil")
print(t)

# Plot data titik
plt.scatter(x, y, marker='*', label='Data Titik')

# Plot hasil interpolasi Lagrange
x1 = np.arange(min(x), max(x) + 0.1, 0.1)
y1 = hasil(x1)
plt.plot(x1, y1, '-o', label=f'Interpolasi Lagrange Pangkat {order_interpolation}')

# Contoh nilai x di luar rentang data
x_outside_range = np.array([0.05,0.02,1.4])
y_outside_range = hasil(x_outside_range)
plt.scatter(x_outside_range, y_outside_range, color='red', marker='x', label='Ekstrapolasi')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('Interpolasi Polinomial Lagrange')
plt.legend()
plt.show()
```

Dengan menggunakan kode pada soal nomor 1 dan ditambahkan dengan nilai yang berada di luar range data. Nilai yang ditambahkan yaitu 0.05, 0.02, dan 1.4. Pada hasil plot, nilai yang ditambahkan ditandai dengan x merah dengan label ekstrapolasi. Ekstrapolasi yaitu memprediksi nilai titik data yang tidak diketahui dengan memproyeksikan di luar rentang titik data yang diketahui. Terlihat titik-titik tersebut berada di luar grafik, sehingga hasilnya kurang akurat.

Apa perbedaan antara regresi linier dan interpolasi Lagrange?

**Regresi linier:**
*	Bertujuan untuk menemukan hubungan linier antara dua variabel (independen dan dependen) atau lebih. Ini digunakan untuk membuat model prediktif yang dapat memprediksi nilai variabel dependen berdasarkan nilai variabel independen.
*	Digunakan ketika kita ingin memahami atau memodelkan hubungan linear antara variabel-variabel. Misalnya, dalam analisis regresi linier sederhana, kita mencoba memprediksi variabel dependen y berdasarkan variabel independen x dengan menggambarkan garis regresi linear.
*	Melibatkan mencari parameter (slope dan intercept) yang menggambarkan garis linear terbaik sesuai dengan data.
*	Digunakan untuk membuat prediksi nilai variabel dependen berdasarkan nilai variabel independen di luar data yang diketahui.

**Interpolasi Lagrange:**
*	Bertujuan untuk membuat polinomial interpolasi yang melewati serangkaian titik data yang diketahui. Interpolasi digunakan untuk mengisi nilai di antara titik-titik data yang diketahui.
*	Digunakan ketika kita memiliki serangkaian titik data dan ingin membuat polinomial yang melalui semua titik tersebut. Ini sering digunakan dalam interpolasi numerik untuk menghitung nilai di antara titik-titik data yang diketahui.
*	Melibatkan pembentukan polinomial interpolasi dari derajat tertentu yang melewati semua titik data yang diketahui.
*	Digunakan untuk mengisi nilai di antara titik-titik data yang diketahui.

Berikan satu contoh kasus biomedis yang dapat menggunakan interpolasi Lagrange untuk menyelesaikannya! Berikan data dan hasil dari interpolasi Lagrange tersebut!
Data Konsentrasi Obat:

|Waktu (jam)	        |0 |	1|	2|	3|	4|
|---------------------|--|---|---|---|---|
|Konsentrasi (mb/mL)	|10|	8|	5|	3|	1|

Kita ingin menghitung konsentrasi obat pada waktu t = 2.5 jam


```python
import numpy as np
import matplotlib.pyplot as plt

def lagrange_interpolation(x, y, xi):
    n = len(x)
    result = 0.0

    for i in range(n):
        term = y[i]
        for j in range(n):
            if i != j:
                term = term * (xi - x[j]) / (x[i] - x[j])
        result += term

    return result

# Data
time = np.array([0, 1, 2, 3, 4])
concentration = np.array([10, 8, 5, 3, 1])

# Waktu yang ingin diinterpolasi
interpolation_time = 2.5

# Menghitung konsentrasi dengan interpolasi Lagrange
interpolated_concentration = lagrange_interpolation(time, concentration, interpolation_time)

# Menampilkan hasil interpolasi
print(f"Konsentrasi obat pada t = {interpolation_time} jam (interpolasi): {interpolated_concentration:.2f} mg/mL")

# Plot data asli dan hasil interpolasi
plt.plot(time, concentration,'-yo', label='Data Asli')
plt.plot(interpolation_time, interpolated_concentration, 'ro', label='Interpolasi Lagrange')
plt.xlabel('Waktu (jam)')
plt.ylabel('Konsentrasi Obat (mg/mL)')
plt.title('Interpolasi Konsentrasi Obat')
plt.legend()
plt.grid(True)
```

Dalam output t	ersebut, hasil interpolasi Lagrange digunakan untuk memperkirakan konsentrasi obat pada waktu 2.5 jam. Dengan menggunakan fungsi interpolasi Lagrange pada data waktu dan konsentrasi yang diberikan, program menghitung nilai konsentrasi obat yang diestimasi pada saat tertentu yang tidak terdapat dalam data asli.
Pada output tabel, nilai-nilai baru dari x (waktu) dan y (konsentrasi obat) ditampilkan beserta hasil interpolasi untuk nilai-nilai tersebut. Misalnya, pada nilai waktu 1jam, konsentrasi obat diestimasi sekitar 9 mg/mL, dan seterusnya. Grafik hasil interpolasi juga ditampilkan, menunjukkan titik-titik interpolasi pada data waktu dan konsentrasi yang diberikan. Didapat konsentrasi obat pada waktu 2.5 jam adalah 3.87 mg/mL yang ditandai dengan titik merah pada grafik.
