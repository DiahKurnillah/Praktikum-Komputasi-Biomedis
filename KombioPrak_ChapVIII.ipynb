## Preliminary Task

Define the analytical value of f '(1) from a function f(x) = x sin(x)!

```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.1
x = np.arange(0, 2*np.pi, h)
y = x * np.sin(x)

def f(x):
    return x * np.sin(x)

forward_diff = (f(x + h) - f(x)) / h
x_diff = x[:]

exact_solution = np.cos(x_diff) + x_diff * np.cos(x_diff)

plt.figure(figsize=(12, 8))
plt.plot(x_diff, forward_diff, '--', label='Finite difference approximation')
plt.plot(x_diff, exact_solution, label='Exact solution')
plt.legend()
plt.show()

max_error = np.max(np.abs(exact_solution - forward_diff))
print('Max Error = ',max_error)
```

Seperti yang ditunjukkan gambar di atas, terdapat offset kecil antara kedua kurva, yang diakibatkan oleh kesalahan numerik dalam evaluasi turunan numerik. Kesalahan maksimal antara dua hasil numerik adalah sekitar 1,25 dan diperkirakan akan berkurang seiring dengan besarnya langkah.

```python
import numpy as np
from scipy.interpolate import lagrange
from prettytable import PrettyTable

# Data
x = np.array([1.2, 1.4, 1.6, 1.8])
y = np.array([0.8333, 0.7143, 0.6250, 0.5556])

# Interpolasi Lagrange
poly = lagrange(x, y)

# Persamaan matematika
equation = np.poly1d(poly)

# Titik evaluasi
x0 = 1.5

# Forward difference approximation
h = 0.1  # Jarak
f_prime_forward = (equation(x0 + h) - equation(x0)) / h

# Backward difference approximation
f_prime_backward = (equation(x0) - equation(x0 - h)) / h

# Central difference approximation
f_prime_central = (equation(x0 + h) - equation(x0 - h)) / (2 * h)

# Menampilkan hasil dalam tabel
t = PrettyTable()
t.field_names = ['Metode', "f'(1.5)"]
t.add_row(['Forward Difference', f_prime_forward])
t.add_row(['Backward Difference', f_prime_backward])
t.add_row(['Central Difference', f_prime_central])

print("Hasil Finite Difference Method:")
print(t)
print(equation)
```

Code yang telah diberikan bertujuan untuk menghitung nilai dari turunan pertama f'(1.5) dari suatu fungsi matematika yang diaproksimasi menggunakan interpolasi Lagrange. Interpolasi Lagrange dilakukan terlebih dahulu untuk mendapatkan persamaan matematika yang menggambarkan hubungan antara variabel x dan y dari data titik yang tersedia.
Langkah pertama adalah melakukan interpolasi Lagrange dengan menggunakan data titik yang diberikan x dan y. Persamaan matematika hasil interpolasi ini disimpan dalam variabel equation. Selanjutnya, kita tentukan titik evaluasi x0 di mana kita ingin menghitung turunan pertama.
Selanjutnya, kita menggunakan metode finite difference approximation untuk menghitung turunan pertama f'(1.5) dengan tiga metode yang berbeda, yaitu forward difference, backward difference, dan central difference. Dalam metode finite difference, kita menggunakan perbedaan nilai fungsi pada titik-titik yang berdekatan dan jarak h yang ditentukan.
Forward Difference menghitung perbedaan nilai fungsi di x0dan x0 + h, kemudian dibagi dengan h. Backward Difference menghitung perbedaan nilai fungsi di x0 dan x0 - h, kemudian dibagi dengan h. Central Difference menghitung perbedaan nilai fungsi di x0 + h dan x0 - h, kemudian dibagi dengan 2h.
