# Preliminary Task
Distribusi panas satu dimensi (1D) sebagai fungsi waktu (t) pada logam memenuhi persamaan berikut:
$$\frac{âˆ‚u}{âˆ‚t} (x,t)-\frac{âˆ‚^2 u}{âˆ‚x^2 } (x,t)=0,    0<x<1     0â‰¤t$$

Dengan syarat batas:

$$u(0,t)=u(1,t)=0,0<t$$,
dan syarat awal:

$$u(x,0) = sin(Ï€x),0â‰¤xâ‰¤1$$,
Selesaikan persamaan parabola PDP tersebut dengan menggunakan metode beda hingga jika sumbu vertikal menunjukkan perubahan dari waktu ke waktu dengan interval k = 0,0005. Karena Î± = 1, h = 0, 1 dan k = 0,0005.


```python
import numpy as np
import matplotlib.pyplot as plt

L = 1.0
T = 0.1
alpha = 1.0
h = 0.1
k = 0.0005

Nx = int(L / h) + 1
Nt = int(T / k) + 1

u = np.zeros((Nx, Nt))
x = np.linspace(0, L, Nx)
u[:, 0] = np.sin(np.pi * x)

for j in range(0, Nt - 1):
    for i in range(1, Nx - 1):
        u[i, j + 1] = u[i, j] + alpha * k / h**2 * (u[i + 1, j] - 2 * u[i, j] + u[i - 1, j])

# Plot hasil dalam grafik dua dimensi
plt.figure(figsize=(10, 6))
plt.imshow(u, aspect='auto', extent=[0, T, 0, L], cmap='hot')
plt.colorbar(label='Temperature')
plt.title('Distribusi Panas pada Batang Logam (2D)')
plt.xlabel('Waktu (s)')
plt.ylabel('Posisi (cm)')
plt.show()

X, T = np.meshgrid(x, np.linspace(0, T, Nt))
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, T, u.T, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('u(x, t)')
ax.set_title('Solusi Distribusi Panas 1D')
plt.show()
```

Gambar yang disajikan merupakan representasi grafis dari distribusi suhu (temperature) di seluruh panjang batang logam selama beberapa waktu t yang berskala linear. Sumbu horizontal (x) menunjukkan posisi dari titik pada batang logam, yang berada antara 0 dan 1 cm. Sumbu vertikal (u) menunjukkan suhu di titik tersebut pada waktu t yang tertentu.
Setiap titik pada garis vertikal menunjukkan nilai suhu di posisi tertentu pada batang logam pada waktu tertentu. Misalnya, titik pada sumbu vertikal sebesar 0,04 pada waktu 0,06 second menunjukkan bahwa suhu di posisi 0,8 cm pada batang logam pada waktu 0,06 second adalah 0,04 derajat Celcius.
Garis horizontal pada gambar menunjukkan posisi batang logam pada waktu tertentu, sedangkan garis vertikal menunjukkan waktu yang diukur. Selain itu, garis horizontal pada bagian bawah gambar menunjukkan posisi batang logam, sementara garis vertikal pada sisi kiri gambar menunjukkan waktu yang diukur.
Hal ini menunjukkan bahwa suhu di ujung-ujung batang logam selalu konstan dan sama dengan nol, sedangkan suhu di tengah batang logam lebih tinggi pada awal, tetapi akan menurun seiring waktu dan akhirnya menjadi nol pada waktu yang cukup lama. Hal ini disebabkan oleh pergerakan panas dari area yang lebih panas ke area yang lebih dingin, sampai akhirnya suhu merata pada seluruh panjang batang logam

Persamaan difusi ditulis dalam satu dimensi spasial sebagai:
$$\frac{âˆ‚u}{âˆ‚t}=D \frac{âˆ‚^2 u}{âˆ‚x^2}$$
Mari kita asumsikan konsentrasi awal:
$$u(x,t_0 )=\frac{1}{Ïƒâˆš2Ï€} exp(-\frac{(x-x_{mean} )^2}{Ïƒ^2} )$$
Dimana ð‘¥ð‘šð‘’ð‘Žð‘› = 0 dan lebar ðœŽ = 0.5. Dengan menggunakan metode beda hingga, plot solusi dari persamaan difusi dalam interval -5 sampai 5 dengan ukuran langkah h = 0.1.


```python
import numpy as np
import matplotlib.pyplot as plt

xmean = 0
Ïƒ = 0.5
L = 5
Nx = 150
Nt = 150
x = np.linspace(-L, L, Nx)
dx = 0.1
D = 1

dt = 0.1 * dx**2 / D
u = (1 / (Ïƒ * np.sqrt(2 * np.pi))) * np.exp((x - xmean)**2 / (-1 * (Ïƒ**2)))

initial = True

plt.plot(x, u, 'k-o')

for i in range(Nt):
    u[1:Nx-1] = u[1:Nx-1] + dt * D * (u[2:Nx] - 2 * u[1:Nx-1] + u[0:Nx-2]) / dx**2
    if initial:
        u[1] = u[0] + dt * D * (u[1] - 2 * u[2] + u[3]) / dx**2
        initial = False

plt.plot(x, u, 'r-o')
plt.xlabel('x')
plt.ylabel('Difusi')
plt.show()
```
